```python
import matplotlib.pyplot as plt
```

```python
df
```

```python
plt.plot(df["PaymentMethod"],df["TotalCharges"])
```

```python
plt.title("tenure vs TotalCharges")
plt.xlabel("tenure")
plt.ylabel("TotalCharges")
plt.show()
```

```python
plt.bar(df["PaymentMethod"],df["TotalCharges"])
plt.title("payment methos vs total charges")
plt.xlabel("payment methos")
plt.ylabel("total charges")
plt.show()
```

```python
plt.scatter(df["PaymentMethod"],df["TotalCharges"])
plt.title("PaymentMethod vs TotalCharges")
plt.show()
```

```python
plt.hist(df["TotalCharges"])
plt.title("PaymentMethod")
plt.show()
```

```python
payment_method_counts = df["PaymentMethod"].value_counts()
plt.pie(payment_method_counts, labels=payment_method_counts.index, autopct='%1.1f%%')
plt.title("Distribution of Payment Methods")
plt.show()
```

```python
plt.plot(df["PaymentMethod"],df["TotalCharges"])
plt.plot(df["PaymentMethod"],df["MonthlyCharges"])
plt.legend(["TotalCharges","MonthlyCharges"])
plt.show()
```
